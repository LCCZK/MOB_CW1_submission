```python
import os
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_1 = pd.read_csv("/home/lukas/MOB/MOB_CW1/pf_analysis/simpath1_mine.csv")
gt_1 = pd.read_csv("/home/lukas/MOB/MOB_CW1/pf_analysis/gt_sim1.csv")

m_x = raw_1[["x value"]].to_numpy()
m_y = raw_1[["value"]].to_numpy()

gt_x = gt_1[["x value"]].to_numpy()
gt_y = gt_1[["value"]].to_numpy()

gt_x = np.add(gt_x,15)
gt_y = np.add(gt_y,15)

plt.plot(gt_x,gt_y, label='/base_pose_ground_truth.position')
plt.plot(m_x,m_y, label='/estimatedpose.position')

plt.legend(loc='best')

plt.title("Simpath1: Estimated position comparered to ground truth")

plt.show()

# est = raw_1.plot(x="x value", y="value")
# gt = gt_1.plot(x="x value", y="value")
# gt.set_xlim(-15,15)
# gt.set_ylim(-15,15)
# plt.plot(raw['x value'], raw['value'], 'bo')
```
